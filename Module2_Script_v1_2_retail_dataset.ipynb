## Clean up the data
Did you notice anything unusual about the "house" amounts? 

How about the handset price and income?

````python
df_clean = df.copy()
````

````python
df_clean['total_amount'] = df['total_amount'].str.replace(',', '').str.replace('$','').astype(float)
````

````python
df_clean['total_amount'] = df['total_amount'].astype(str).str.replace(',', '').str.replace('$','').astype(float)

````

````python
df_clean['product_price'] = df['product_price'].str.replace(',', '').str.replace('$','').astype(float)
````

````python
df_clean.describe()
````

````python
df_clean.dropna(inplace=True)
````

````python
df_clean.groupby(['order_year_month']).count()
````
